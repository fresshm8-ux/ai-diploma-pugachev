# Занятие 3. INSERT, UPDATE и DELETE

## Цель занятия

На прошлом занятии мы научились получать данные через `SELECT`.

Сегодня изучаем:
- `INSERT` — добавление данных;
- `UPDATE` — изменение данных;
- `DELETE` — удаление данных.

Это основные операции изменения базы данных.

---

## Что такое CRUD

Практически любое приложение работает с 4 основными операциями:

| Операция | SQL |
|---|---|
| Create | INSERT |
| Read | SELECT |
| Update | UPDATE |
| Delete | DELETE |

Это называется CRUD.

---

## Где это используется в реальной жизни

Примеры:

- HR-система добавляет новое резюме;
- интернет-магазин обновляет цену товара;
- спортивное приложение изменяет тренировку;
- система питания удаляет старое меню;
- Telegram добавляет новые сообщения.

---

## Что такое INSERT

`INSERT INTO` добавляет новые записи.

Пример:

```sql
INSERT INTO students (name, age)
VALUES ('Анна', 20)
```

---

## Что такое UPDATE

`UPDATE` изменяет существующие данные.

Пример:

```sql
UPDATE students
SET age = 21
WHERE name = 'Анна'
```

---

## Что такое DELETE

`DELETE` удаляет записи.

Пример:

```sql
DELETE FROM students
WHERE id = 1
```

---

## Почему WHERE очень важен

Если забыть `WHERE`, можно изменить или удалить всю таблицу.

Опасный запрос:

```sql
DELETE FROM students
```

Он удалит ВСЕ записи.

---

## Зачем это нужно в дипломе

В дипломном проекте пользователь почти всегда:
- создаёт данные;
- изменяет данные;
- удаляет данные.

Например:
- HR-проект обновляет статус кандидата;
- питание обновляет калории;
- тренировки обновляют длительность;
- магазин обновляет цены.

## Ячейка 1. Подключаем sqlite3 и создаём базу

````python
import sqlite3

connection = sqlite3.connect("crud_lesson03.db")
cursor = connection.cursor()

print("База подключена")
````

## Ячейка 2. Создаём таблицу employees

````python
cursor.execute("""
CREATE TABLE IF NOT EXISTS employees (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    position TEXT,
    salary INTEGER
)
""")

connection.commit()

print("Таблица employees создана")
````

## Ячейка 3. Очищаем таблицу

````python
cursor.execute("DELETE FROM employees")

connection.commit()

print("Таблица очищена")
````

## Ячейка 4. INSERT — добавляем сотрудников

````python
employees = [
    ("Анна", "HR", 120000),
    ("Иван", "Developer", 200000),
    ("Ольга", "Analyst", 150000)
]

cursor.executemany(
    "INSERT INTO employees (name, position, salary) VALUES (?, ?, ?)",
    employees
)

connection.commit()

print("Сотрудники добавлены")
````

## Ячейка 5. Проверяем данные через SELECT

````python
cursor.execute("SELECT * FROM employees")

employees_data = cursor.fetchall()

for employee in employees_data:
    print(employee)

assert len(employees_data) == 3
````

## Ячейка 6. UPDATE — изменяем зарплату

````python
cursor.execute(
    "UPDATE employees SET salary = ? WHERE name = ?",
    (220000, "Иван")
)

connection.commit()

print("Зарплата обновлена")
````

## Ячейка 7. Проверяем UPDATE

````python
cursor.execute(
    "SELECT * FROM employees WHERE name = ?",
    ("Иван",)
)

ivan = cursor.fetchone()

print(ivan)

assert ivan[3] == 220000
````

## Ячейка 8. DELETE — удаляем запись

````python
cursor.execute(
    "DELETE FROM employees WHERE name = ?",
    ("Ольга",)
)

connection.commit()

print("Запись удалена")
````

## Ячейка 9. Проверяем DELETE

````python
cursor.execute("SELECT * FROM employees")

employees_after_delete = cursor.fetchall()

for employee in employees_after_delete:
    print(employee)

assert len(employees_after_delete) == 2
````

## Ячейка 10. Закрываем соединение

````python
connection.close()

print("Соединение закрыто")
````
