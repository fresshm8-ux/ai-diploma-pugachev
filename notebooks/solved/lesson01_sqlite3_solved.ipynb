---

## Что такое база данных

**База данных** — это организованное хранилище информации.

Примеры из реальной жизни:

- интернет-магазин хранит товары, покупателей и заказы;
- HR-система хранит резюме кандидатов;
- приложение питания хранит продукты, калории и приёмы пищи;
- спортивное приложение хранит тренировки и результаты;
- учебная система хранит студентов, задания и оценки.

---

## Что такое таблица

В реляционной базе данных информация хранится в таблицах.

Пример таблицы `students`:

| id | name | city | age |
|---|---|---|---|
| 1 | Анна | Казань | 20 |
| 2 | Иван | Москва | 21 |

- строка — одна запись;
- столбец — один признак;
- `id` — уникальный номер записи.

---

## Что такое SQL

**SQL** — язык запросов к базе данных.

Сегодня используем первые команды:

- `CREATE TABLE` — создать таблицу;
- `INSERT INTO` — добавить данные;
- `SELECT` — получить данные;
- `WHERE` — отфильтровать данные.

---

## Что такое sqlite3

`sqlite3` — встроенный модуль Python для работы с SQLite-базами данных.

SQLite удобна для обучения, потому что:

- не требует установки сервера;
- база хранится в одном файле;
- работает прямо в Google Colab;
- подходит для учебных и небольших дипломных проектов.

---

## Зачем это нужно в дипломе

В дипломном проекте база данных может хранить:

- пользователей;
- товары;
- резюме;
- тренировки;
- продукты питания;
- результаты анализа;
- историю действий пользователя.

Даже простой диплом выглядит сильнее, если данные хранятся не в переменных, а в базе данных.

## Ячейка 1. Подключаем sqlite3

`sqlite3` уже встроен в Python.  
В Google Colab его не нужно устанавливать отдельно.

````python
import sqlite3

print("sqlite3 успешно подключен")
````

## Ячейка 2. Создаём файл базы данных

Команда `sqlite3.connect(...)` создаёт файл базы данных, если его ещё нет.

Здесь мы создаём учебную базу:

`students_lesson01.db`

````python
connection = sqlite3.connect("students_lesson01.db")

print("База данных создана или открыта")
````

## Ячейка 3. Создаём cursor

`cursor` — это объект, через который Python отправляет SQL-запросы в базу данных.

````python
cursor = connection.cursor()

print("Cursor создан")
````

## Ячейка 4. Создаём таблицу students

Используем SQL-команду `CREATE TABLE`.

Поля таблицы:

- `id` — уникальный номер;
- `name` — имя студента;
- `city` — город;
- `age` — возраст.

````python
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    city TEXT,
    age INTEGER
)
""")

print("Таблица students создана")
````

## Ячейка 5. Добавляем данные в таблицу

Используем SQL-команду `INSERT INTO`.

Знак `?` — безопасная подстановка значения.  
Такой способ защищает программу от ошибок и SQL-инъекций.

````python
cursor.execute(
    "INSERT INTO students (name, city, age) VALUES (?, ?, ?)",
    ("Анна", "Казань", 20)
)

cursor.execute(
    "INSERT INTO students (name, city, age) VALUES (?, ?, ?)",
    ("Иван", "Москва", 21)
)

cursor.execute(
    "INSERT INTO students (name, city, age) VALUES (?, ?, ?)",
    ("Ольга", "Казань", 19)
)

print("Данные добавлены")
````

## Ячейка 6. Сохраняем изменения

`commit()` сохраняет изменения в базе данных.

Без `commit()` новые записи могут не сохраниться.

````python
connection.commit()

print("Изменения сохранены")
````

## Ячейка 7. Получаем все данные

Используем SQL-команду:

```sql
SELECT * FROM students
```

Она означает: получить все строки и все столбцы из таблицы `students`.

````python
cursor.execute("SELECT * FROM students")

students = cursor.fetchall()

print(students)

assert len(students) >= 3
````

## Ячейка 8. Красиво выводим данные

Теперь выводим строки таблицы по одной, чтобы результат было легче читать.

````python
for student in students:
    print("ID:", student[0], "| Имя:", student[1], "| Город:", student[2], "| Возраст:", student[3])
````

## Ячейка 9. Фильтруем данные через WHERE

Получим только студентов из Казани.

SQL-запрос:

```sql
SELECT * FROM students WHERE city = ?
```

````python
cursor.execute(
    "SELECT * FROM students WHERE city = ?",
    ("Казань",)
)

kazan_students = cursor.fetchall()

print(kazan_students)

assert len(kazan_students) >= 1
````

## Ячейка 10. Закрываем соединение

После работы с базой соединение нужно закрывать.

````python
connection.close()

print("Соединение с базой данных закрыто")
````
