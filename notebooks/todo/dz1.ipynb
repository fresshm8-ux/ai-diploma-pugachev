```python
#1
import sqlite3

print("sqlite3 успешно подключен")
```

```python
#2
connection = sqlite3.connect("students_lesson01.db")

print("База данных создана или открыта")
```

```python
#3
cursor = connection.cursor()

print("Cursor создан")
```

```python
#4
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    city TEXT,
    age INTEGER
)
""")

print("Таблица students создана")
```

```python
#5
cursor.execute(
    "INSERT INTO students (name, city, age) VALUES (?, ?, ?)",
    ("Анна", "Казань", 20)
)

cursor.execute(
    "INSERT INTO students (name, city, age) VALUES (?, ?, ?)",
    ("Иван", "Москва", 21)
)

cursor.execute(
    "INSERT INTO students (name, city, age) VALUES (?, ?, ?)",
    ("Ольга", "Казань", 19)
)

cursor.execute(
    "INSERT INTO students (name, city, age) VALUES (?, ?, ?)",
    ("Петр", "Санкт-Петербург", 22)
)

cursor.execute(
    "INSERT INTO students (name, city, age) VALUES (?, ?, ?)",
    ("Мария", "Казань", 20)
)

print("Данные добавлены")
```

```python
#6
connection.commit()

print("Изменения сохранены")
```

```python
#7 SELECT * FROM students
cursor.execute("SELECT * FROM students")

students = cursor.fetchall()

print(students)

assert len(students) >= 3
assert len(students) >= 4
assert len(students) >= 5
```

```python
#8
for student in students:
    print("ID:", student[0], "| Имя:", student[1], "| Город:", student[2], "| Возраст:", student[3])
```
