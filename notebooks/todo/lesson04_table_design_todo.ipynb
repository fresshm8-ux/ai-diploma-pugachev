## Зачем проектировать таблицы

Таблица — это не просто набор столбцов.

Если таблица плохо спроектирована, потом будет сложно:
- добавлять данные;
- искать данные;
- фильтровать данные;
- делать аналитику;
- развивать дипломный проект.

Пример плохой таблицы:

| data |
|---|
| Анна, Казань, 20, студент |

Всё лежит в одном столбце. Такую таблицу неудобно анализировать.

Пример хорошей таблицы:

| id | name | city | age | role |
|---|---|---|---|---|
| 1 | Анна | Казань | 20 | student |

Каждый признак лежит в отдельном столбце.

---

## Основные типы данных в SQLite

| Тип | Что хранит |
|---|---|
| INTEGER | целые числа |
| TEXT | текст |
| REAL | числа с точкой |
| NULL | пустое значение |
| BLOB | бинарные данные |

В этом курсе чаще всего используем:
- `INTEGER`;
- `TEXT`;
- `REAL`.

---

## Что такое PRIMARY KEY

`PRIMARY KEY` — это главный уникальный идентификатор записи.

Обычно это поле `id`.

Пример:

```sql
id INTEGER PRIMARY KEY AUTOINCREMENT
```

Это означает:
- `id` — номер записи;
- `PRIMARY KEY` — уникальный ключ;
- `AUTOINCREMENT` — номер увеличивается автоматически.

---

## Почему это важно в реальной жизни

В реальных системах почти у каждой записи есть уникальный номер:

- номер заказа;
- номер пользователя;
- номер резюме;
- номер тренировки;
- номер товара.

Без `id` сложно точно понимать, какую запись нужно изменить или удалить.
