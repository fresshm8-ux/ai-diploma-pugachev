```python
#1
import sqlite3

print("sqlite3 подключен")
```

```python
#2
def get_connection(db_name="python_sql_lesson08.db"):
    connection = sqlite3.connect(db_name)
    return connection

connection = get_connection()

print("Подключение создано")

assert connection is not None
```

```python
#3
def create_candidates_table(connection):
    cursor = connection.cursor()
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS candidates (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    city TEXT,
    age INTEGER,
    rating INTEGER,
    revenue INTEGER
    )
    """)
    connection.commit()

create_candidates_table(connection)

print("Таблица candidates создана")
```

```python
#4
def clear_candidates(connection):
    cursor = connection.cursor()
    cursor.execute("DELETE FROM candidates")
    connection.commit()

clear_candidates(connection)

print("Таблица candidates очищена")
```

```python
#5
def add_candidate(connection, name, city, age, rating, revenue):
    cursor = connection.cursor()
    cursor.execute(
        "INSERT INTO candidates (name, city, age, rating, revenue) VALUES (?, ?, ?, ?, ?)",
        (name, city, age, rating, revenue)
    )
    connection.commit()

add_candidate(connection, "Анна", "Казань", 20, 5, 1200)
add_candidate(connection, "Иван", "Москва", 21, 4, 950)
add_candidate(connection, "Ольга", "Казань", 19, 3, 700)

print("Кандидаты добавлены")
```

```python
#6
def get_all_candidates(connection):
    cursor = connection.cursor()
    cursor.execute("SELECT * FROM candidates")
    return cursor.fetchall()

candidates = get_all_candidates(connection)

for candidate in candidates:
    print(candidate)

assert len(candidates) == 3
```

```python
#7
def find_candidates_by_city(connection, city):
    cursor = connection.cursor()
    cursor.execute(
        "SELECT * FROM candidates WHERE city = ?",
        (city,)
    )
    return cursor.fetchall()

kazan_candidates = find_candidates_by_city(connection, "Казань")

for candidate in kazan_candidates:
    print(candidate)

assert len(kazan_candidates) == 2
```

```python
#8
def find_candidate_by_name(connection, name):
    cursor = connection.cursor()
    cursor.execute(
        "SELECT * FROM candidates WHERE name = ?",
        (name,)
    )
    return cursor.fetchone()

anna = find_candidate_by_name(connection, "Анна")

print(anna)

assert anna[1] == "Анна"
```

```python
#9
def update_candidate_age(connection, name, new_age):
    cursor = connection.cursor()
    cursor.execute(
        "UPDATE candidates SET age = ? WHERE name = ?",
        (new_age, name)
    )
    connection.commit()

update_candidate_age(connection, "Анна", 22)

anna_updated = find_candidate_by_name(connection, "Анна")

print(anna_updated)

assert anna_updated[3] == 22
```
