---

## Что такое CRUD

Практически любое приложение работает с 4 основными операциями:

| Операция | SQL |
|---|---|
| Create | INSERT |
| Read | SELECT |
| Update | UPDATE |
| Delete | DELETE |

Это называется CRUD.

---

## Где это используется в реальной жизни

Примеры:

- HR-система добавляет новое резюме;
- интернет-магазин обновляет цену товара;
- спортивное приложение изменяет тренировку;
- система питания удаляет старое меню;
- Telegram добавляет новые сообщения.

---

## Что такое INSERT

`INSERT INTO` добавляет новые записи.

Пример:

```sql
INSERT INTO students (name, age)
VALUES ('Анна', 20)
```

---

## Что такое UPDATE

`UPDATE` изменяет существующие данные.

Пример:

```sql
UPDATE students
SET age = 21
WHERE name = 'Анна'
```

---

## Что такое DELETE

`DELETE` удаляет записи.

Пример:

```sql
DELETE FROM students
WHERE id = 1
```

---

## Почему WHERE очень важен

Если забыть `WHERE`, можно изменить или удалить всю таблицу.

Опасный запрос:

```sql
DELETE FROM students
```

Он удалит ВСЕ записи.

---

## Зачем это нужно в дипломе

В дипломном проекте пользователь почти всегда:
- создаёт данные;
- изменяет данные;
- удаляет данные.

Например:
- HR-проект обновляет статус кандидата;
- питание обновляет калории;
- тренировки обновляют длительность;
- магазин обновляет цены.
