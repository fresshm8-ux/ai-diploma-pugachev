```python
#1

import sqlite3

connection = sqlite3.connect("crud_lesson03.db")
cursor = connection.cursor()

print("База подключена")
```

```python
#2
cursor.execute("""
CREATE TABLE IF NOT EXISTS products (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    weight TEXT,
    calories INTEGER
)
""")

connection.commit()

print("Таблица products создана")
```

```python
#3
cursor.execute("DELETE FROM products")

connection.commit()

print("Таблица очищена")
```

```python
#4

products = [
    ("Hamburger", "150g", 300),
    ("Cheeseburger", "180g", 350),
    ("Fries", "120g", 250),
    ("Cola", "300ml", 150)
]

cursor.executemany(
    "INSERT INTO products (name, weight, calories) VALUES (?, ?, ?)",
    products
)

connection.commit()

print("Продукты добавлены")
```

```python
#5
cursor.execute("SELECT * FROM products")

products_data = cursor.fetchall()

for product in products_data:
    print(product)

assert len(products_data) == 4
```

```python
#6
cursor.execute(
    "UPDATE products SET calories = ? WHERE name = ?",
    (350, "Hamburger")
)

connection.commit()

print("Калории обновлена")
```

```python
#7

cursor.execute(
    "SELECT * FROM products WHERE name = ?",
    ("Hamburger",)
)

Hamburger = cursor.fetchone()

print(Hamburger)

assert Hamburger[3] == 350
```

```python
#8
cursor.execute(
    "DELETE FROM products WHERE name = ?",
    ("Cola",)
)

connection.commit()

print("Запись удалена")
```

```python
#9
cursor.execute("SELECT * FROM products")

products_after_delete = cursor.fetchall()

for employee in products_after_delete:
    print(employee)

assert len(products_after_delete) == 3
```

```python
#10
connection.close()

print("Соединение закрыто")
```
