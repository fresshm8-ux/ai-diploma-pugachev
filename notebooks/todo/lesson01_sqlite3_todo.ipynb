## Что такое база данных

**База данных** — это организованное хранилище информации.

Примеры из реальной жизни:

- интернет-магазин хранит товары, покупателей и заказы;
- HR-система хранит резюме кандидатов;
- приложение питания хранит продукты, калории и приёмы пищи;
- спортивное приложение хранит тренировки и результаты;
- учебная система хранит студентов, задания и оценки.

---

## Что такое таблица

В реляционной базе данных информация хранится в таблицах.

Пример таблицы `students`:

| id | name | city | age |
|---|---|---|---|
| 1 | Анна | Казань | 20 |
| 2 | Иван | Москва | 21 |

- строка — одна запись;
- столбец — один признак;
- `id` — уникальный номер записи.

---

## Что такое SQL

**SQL** — язык запросов к базе данных.

Сегодня используем первые команды:

- `CREATE TABLE` — создать таблицу;
- `INSERT INTO` — добавить данные;
- `SELECT` — получить данные;
- `WHERE` — отфильтровать данные.

---

## Что такое sqlite3

`sqlite3` — встроенный модуль Python для работы с SQLite-базами данных.

SQLite удобна для обучения, потому что:

- не требует установки сервера;
- база хранится в одном файле;
- работает прямо в Google Colab;
- подходит для учебных и небольших дипломных проектов.

---

## Зачем это нужно в дипломе

В дипломном проекте база данных может хранить:

- пользователей;
- товары;
- резюме;
- тренировки;
- продукты питания;
- результаты анализа;
- историю действий пользователя.

Даже простой диплом выглядит сильнее, если данные хранятся не в переменных, а в базе данных.
